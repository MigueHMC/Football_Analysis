```python
import numpy as np
import pandas as pd
```

```python
time = pd.date_range('2024-01-01', periods=100)
data = np.random.randn(100).cumsum()
```

```python
df = pd.DataFrame(data, index=time, columns=['value'])
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df.index, df['value'])
```

```python
plt.scatter(df.index, df['value'])
```

```python
import seaborn as sns
```

```python
sns.lineplot(data=df, x=df.index,y='value')
```
