```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.scrapethissite.com/pages/simple/'
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
title = soup.select_one('h1').text
print(title)
```

```python
descripition = soup.select_one('p[class="lead"]').text
print(descripition)
```

```python
country_names = soup.select('h3[class="country-name"]')
country_names = [x.text.strip() for x in country_names]
print(country_names)
```

```python
wild_names = soup.select('h3[class*="-name"]')
wild_names = [x.text.strip() for x in wild_names]
print(wild_names)
```

```python
andorra = soup.select_one('h3:-soup-contains("Andorra")').text
print(andorra.strip())
```

```python
capitals = soup.select('span[class="country-capital"]')
print([x.text.strip() for x in capitals])
```
